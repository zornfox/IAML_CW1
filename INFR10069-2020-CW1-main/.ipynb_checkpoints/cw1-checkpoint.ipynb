```python
w1 = np.linspace(-2,2, 100)
mse_array = []
for i in range(len(w1)):
    W = np.array([20,w1[i]])
    new_predict = X.dot(W)
    W_mse =mean_squared_error(y, new_predict)
    mse_array.append(W_mse)
    
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(w1, mse_array, color='blue', linewidth=3)
ax.set(title = "MSE giving different w1", xlabel = 'w1', ylabel = 'MSE')
plt.savefig('1_g.png', format="png")
plt.show()
```

```python
data_path_q2 = os.path.join(os.getcwd(), 'data', 'regression_part2.csv')
q2 = pd.read_csv(data_path_q2, delimiter = ',')
q2.head(5)
```

```python
q2.info()
```

```python

X=np.array(q2.iloc[:, 0:1]).reshape(-1,1)
X
```

```python
y=np.array(q2.iloc[:,1]).reshape(-1,1)
y
```

```python
import operator
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(np.array(q2.iloc[:,0]).reshape(-1,1), y,  color='black', label='training data')
ax.set(title = "output vs input in 4 different polynomial regression models", xlabel = 'input', ylabel = 'output')
for m in range(1, 5, 1):
    print(m)
    poly = PolynomialFeatures(degree = m)
    X_poly = poly.fit_transform(X)
    model = LinearRegression(fit_intercept = False)
    model.fit(X_poly, y)
    print(model.coef_)
    predicted = model.predict(X_poly)
    sort_axis = operator.itemgetter(0)
    sorted_zip = sorted(zip(X,predicted), key=sort_axis)
    x, y_poly_pred = zip(*sorted_zip)
    ax.plot(x, y_poly_pred, label = "M={}".format(m))
plt.legend()
#plt.savefig('2_a.png', format="png")
plt.show()
```

```python
MSE = []
bars = ('M=1', 'M=2', 'M=3', 'M=4')
y_pos = np.arange(len(bars))
for i in range(1, 5, 1):
    poly = PolynomialFeatures(degree = i)
    X_poly = poly.fit_transform(X)
    poly.fit(X_poly, y)
    model = LinearRegression(fit_intercept = False)
    model.fit(X_poly, y)
    predicted = model.predict(X_poly)
    MSE.append(mean_squared_error(y, predicted))    
# Create bars
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(y_pos, MSE)

# Assign names on the x-axis
plt.xticks(y_pos, bars)
ax.set(title = "MSE for 4 different polynomial regression models", xlabel = 'degree of polynomial features', ylabel = 'MSE')

#plt.savefig('2_b.png', format="png")
plt.show()
```

```python
MSE_3_4 = []
for i in range(3, 5, 1):
    poly = PolynomialFeatures(degree = i)
    X_poly = poly.fit_transform(X)
    poly.fit(X_poly, y)
    model = LinearRegression(fit_intercept = False)
    model.fit(X_poly, y)
    predicted = model.predict(X_poly)
    MSE_3_4.append(mean_squared_error(y, predicted)) 
MSE_3_4
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X, y, color='black', label='Dataset')
ax.set(title = "3 Nonlinear prediction models with RBF kernel in different \u03B1", xlabel = 'input', ylabel = 'output')
alpha = [0.2, 100, 1000]
c = [-4.0, -2.0, 2.0, 4.0]
for a in alpha:
    q2['pre'] = 1
    q2['r1'] = np.exp(-0.5*(q2['input']-c[0])**2/a**2)
    q2['r2'] = np.exp(-0.5*(q2['input']-c[1])**2/a**2)
    q2['r3'] = np.exp(-0.5*(q2['input']-c[2])**2/a**2)
    q2['r4'] = np.exp(-0.5*(q2['input']-c[3])**2/a**2)
    x = q2[['pre','r1', 'r2', 'r3', 'r4']]
    y = q2['output']
    model = LinearRegression()
    model.fit(x, y)
    predicted = model.predict(x)
    sort_axis = operator.itemgetter(0)
    sorted_zip = sorted(zip(X,predicted), key=sort_axis)
    x, y_poly_pred = zip(*sorted_zip)
    ax.plot(x, y_poly_pred, label = "\u03B1={}".format(a))

plt.legend()
#plt.savefig('2_d.png')
plt.show()
```
